# Data preprocessing

Based on preprocessing designed by **Ravi Chaubey** on [Kaggle](https://www.kaggle.com/code/ravichaubey1506/end-to-end-machine-learning/notebook).

```python
import pandas as pd
import os

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
os.chdir("..")
```

```python
df = pd.read_csv("resources/data/housing.csv")
df.head()
```

```python
X, y = df.drop(columns=["median_house_value"]), df[["median_house_value"]]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
```

```python
num_pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy="median")),
        ("std_scaler", StandardScaler()),
    ]
)

housing_num = df.drop(columns=["ocean_proximity"], axis=1)

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]
```

```python
full_pipeline = ColumnTransformer(
    [
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ]
)
```

```python
labels = [
    "longitude",
    "latitude",
    "housing_median_age",
    "total_rooms",
    "total_bedrooms",
    "population",
    "households",
    "median_income",
    "median_house_value",
    "ocean_proximity_<1H OCEAN",
    "ocean_proximity_INLAND",
    "ocean_proximity_ISLAND",
    "ocean_proximity_NEAR BAY",
    "ocean_proximity_NEAR OCEAN",
]
```

```python
df_preproc = full_pipeline.fit_transform(df)
df_preproc = pd.DataFrame(data=df_preproc, columns=labels)

df_preproc.head()
```

```python
pd.to_pickle(df_preproc, "resources/data/housing_preproc.pkl")
```
