# Model summary

```python
import os
import sys

os.chdir('../')

sys.path.append('..')
```

```python
import pandas as pd
import dalex as dx
import pickle as pkl

from pipelines.config import train_data_full_path, test_data_full_path, model_path
```

```python
df_test, df_train = pd.read_csv(test_data_full_path), pd.read_csv(train_data_full_path)
X_test, y_test = df_test.drop(columns=['median_house_value']), df_test[['median_house_value']]
X_train, y_train = df_train.drop(columns=['median_house_value']), df_train[['median_house_value']]
```

```python
models_paths = [f for f in os.listdir('resources/models')]
```

```python
def evaluate_model(model, name: str, X: pd.DataFrame, y: pd.DataFrame) -> pd.DataFrame:
    exp = dx.Explainer(model, X, y, label=name)
    return exp.model_performance().result
```

```python
%%capture

models_df = pd.DataFrame()

for path in models_paths:
    with open(os.path.join(model_path, path), 'rb') as f:
        model = pkl.load(f)
    name = os.path.splitext(path)[0]
    models_df = models_df.append(evaluate_model(model, name, X_test, y_test), ignore_index=False)
```

```python
models_df.sort_values('r2', ascending=False).sort_index('index', key=lambda x:x.map(lambda y: y[0]), kind='stable')
```
